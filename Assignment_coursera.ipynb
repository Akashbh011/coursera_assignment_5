
**Question 1 - Extracting Tesla Stock Data Using yfinance**

```python
# Install yfinance if not already installed


import yfinance as yf

# Define the ticker symbol for Tesla
tesla = yf.Ticker("TSLA")

# Fetch Tesla stock data
tesla_stock_data = tesla.history(period="1y")

# Display the first few rows of the data
tesla_stock_data.head()
```


**Question 2 - Extracting Tesla Revenue Data Using Webscraping**

```python
from bs4 import BeautifulSoup
import requests

# URL for Tesla's revenue data (example)
url = 'https://stockanalysis.com/quote/neo/TSLA/revenue/'  # Replace with the actual URL

# Request the page
response = requests.get(url)

# Parse the page content
soup = BeautifulSoup(response.content, 'html.parser')

# Extract revenue data (modify selectors as needed)
revenue_dataa = soup.find('div', class_='flex-1').text  # Adjust based on actual HTML

# Display the revenue data
print("Tesla Revenue Data:", revenue_dataa)
```

**Extracting GameStop Stock Data Using yfinance**

```python
# Define the ticker symbol for GameStop
gamestop = yf.Ticker("GME")

# Fetch GameStop stock data
gamestop_stock_data = gamestop.history(period="1y")

# Display the first few rows of the data
gamestop_stock_data.head()
```

**Question 4 - Extracting GameStop Revenue Data Using Webscraping**

```python
# URL for GameStop's revenue data (example)
from bs4 import BeautifulSoup
import requests

# URL for GameStop's revenue data (example)
url = 'https://stockanalysis.com/stocks/gme/revenue/'
# Request the page
response = requests.get(url)

# Parse the page content
soup = BeautifulSoup(response.content, 'html.parser')
# <div class="flex-1 sm:ml-3 md:flex md:justify-between">
# Extract revenue data from a <span> tag with a specific style
# Adjust the style value to match the one used in the HTML of the page
span_tag = soup.find('div', class_='flex-1 sm:ml-3 md:flex md:justify-between')  # Example style

if span_tag:
    revenue_data = span_tag.text
else:
    revenue_data = 'Revenue data not found'

# Display the revenue data
print("GameStop Revenue Data:", revenue_data)

```

**Question 5 - Tesla Stock and Revenue Dashboard**

```python
import matplotlib.pyplot as plt

# Plot Tesla stock data
plt.figure(figsize=(12, 6))
plt.plot(tesla_stock_data.index, tesla_stock_data['Close'], label='Closing Price')
plt.title('Tesla Stock Closing Price')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()

# Display Tesla revenue data
print("Tesla Revenue Data:", revenue_dataa)  # Ensure `revenue_data` from Question 2 is available
```
